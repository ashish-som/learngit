```python
# Let's cap the outliers using IQR method
```

```python
def out_cap(col):
    q3=num[col].quantile(.75)
    q1=num[col].quantile(.25)
    
    iqr=q3-q1
    
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    
    num[col].clip(lower,upper,inplace=True)
```

```python
for i in cols:
    out_cap(i)
```
